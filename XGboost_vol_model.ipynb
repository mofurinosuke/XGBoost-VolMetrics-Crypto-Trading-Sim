###ローソク足データとは

アルゴリズムトレードモデルは、単一の時点の価格だけでなく、一定期間の価格変動パターンを分析する必要があります。そこで用いたのが「ローソク足データ」です。

ローソク足データは、特定の時間枠（例：1分、1時間、1日）における価格の動きを、一本のローソクの形で視覚的に表現したものです。一本のローソク足には、以下の4つの重要な情報が詰まっています。

* 始値（Open）: その時間枠の最初の取引価格

* 高値（High）: その時間枠の最高価格

* 安値（Low）: その時間枠の最低価格

* 終値（Close）: その時間枠の最後の取引価格

ローソク足の「胴体」の色（緑や赤など）や長さは、価格がその時間枠でどれだけ変動したかを示し、「ヒゲ」の長さは最高値と最低値の幅を表します。このデータを分析することで、市場のトレンドやボラティリティを詳細に把握し、モデルの予測精度を高めることができました。

###データ整形
CSVファイルに別々に保存した、ETH_JPYとBTC_JPYの4本足データを時間がずれないように一つにまとめ、df_rawに代入します。



````python
import pandas as pd

def load_and_align(csv_eth, csv_btc):
    # 1. CSV読み込み
    eth = pd.read_csv(csv_eth, parse_dates=["timestamp"])
    btc = pd.read_csv(csv_btc, parse_dates=["timestamp"])

    # 2. timestampをindex化
    eth = eth.set_index("timestamp")
    btc = btc.set_index("timestamp")

    # 3. 時間足を1時間に揃える（欠損を埋めるためにリサンプル）
    # →ここではlast価格を残す。volumeはsumでもOK。
    eth = eth.resample("1H").agg({
        "open": "first",
        "high": "max",
        "low": "min",
        "close": "last",
        "volume": "sum"
    })
    btc = btc.resample("1H").agg({
        "open": "first",
        "high": "max",
        "low": "min",
        "close": "last",
        "volume": "sum"
    })

    # 4. inner joinで共通の時間帯だけ残す
    aligned = eth.join(btc, how="inner", lsuffix="_eth", rsuffix="_btc")

    return aligned

eth = "/content/eth_jpy_data_2021_2025_1h.csv"
btc ="/content/btc_jpy_data_2017_2025_1h.csv"
df_raw = load_and_align(eth, btc)
````

## GARCH(1,1)モデルによる条件付きボラティリティの推論

このモデルの核となる特徴量の一つが、`calculate_garch_volatility` 関数で計算される **GARCH(1,1)モデルに基づくボラティリティ**です。

### 1. なぜGARCHが必要なのか

金融時系列データ（リターン）は、**ボラティリティのクラスタリング**という特性を持つことが知られています。これは「大きな変動の後には大きな変動が続き、小さな変動の後には小さな変動が続く」という現象です。

* **単純な標準偏差の限界**: 過去$N$期間の単純な標準偏差（例：移動平均標準偏差）は、過去の全てのデータに同じ重みをかけ、この**時間的な変動の連鎖**を捉えることができません。
* **GARCHの優位性**: **GARCH** モデルは、現在のボラティリティを**過去のリターンの大きさ**と**過去の予測ボラティリティ自体**に依存すると仮定し、このクラスタリング現象を確率論的にモデル化します。

### 2. GARCH(1,1)とは

GARCH(1,1)は、**条件付き分散**（$\sigma_t^2$）が以下の3つの要素で決定されるとモデル化します。

$$\sigma_t^2 = \omega + \alpha \epsilon_{t-1}^2 + \beta \sigma_{t-1}^2$$

* $\sigma_t^2$: 時刻 $t$ における**将来のボラティリティ（分散）の予測値**。
* **$\omega$ (定数項)**: 市場の**長期的な平均ボラティリティ**への回帰力を表します。
* **$\alpha \epsilon_{t-1}^2$ (ARCH項, $\alpha$ > 0)**: 直前の期間のリターン $\epsilon_{t-1}$ の**衝撃（二乗）**が、現在のボラティリティに与える影響。直近の大きなニュースや急激なイベントによる**即時的な変動**を捉えます。
* **$\beta \sigma_{t-1}^2$ (GARCH項, $\beta$ > 0)**: 直前の期間の**予測ボラティリティ自体**が、現在のボラティリティに与える影響。ボラティリティの**慣性や持続性**（記憶）を捉える役割を果たします。

**【数式の意味】**：GARCH(1,1)を用いることで作られる特徴量は単なる過去の測定値ではなく、「**平均へ回帰する力**」と「**直近の変動**」、そして「**ボラティリティの持続的な慣性**」の三つの要素を考慮に入れた次期のリスク指標となります。例えば「**このリスク指標の大きさと同じだけ価格が下がった場合、次の期間で価格が一定値上がる確率**」、などといった新たな条件付けと結果の検証を行うことができるようになります。また、aiに直接学習させることで、予測精度を向上させることができるかもしれません。

### 3. 実装上の工夫

* **ローリングウィンドウ**: `window` サイズ（デフォルト100期間）の**ローリングウィンドウ**を使用し、**モデルパラメータの非定常性**に対応。市場環境が変化するたびにモデルを再適合させることで、**最新の市場構造**を反映した予測を試みています。
* **エラー処理**: GARCHモデルは収束しない場合があるため、`try-except`ブロックを設け、堅牢な計算処理を実現しています。

## 🧭 ADX (Average Directional Index) による市場トレンドの定量化

`add_adx` 関数で計算される **ADX (Average Directional Index)** は、古典的なテクニカル分析指標ですが、機械学習モデルの特徴量として組み込むことで、**トレンド環境**と**レンジ環境**を明確に区別する重要なメタ情報を提供します。

### 1. ADXを採用する理論的根拠（市場環境の分類）

多くの価格予測モデル、特にブレイクアウト戦略や順張り戦略は、**強いトレンド環境**でのみ高い性能を発揮します。逆に、価格が一定の範囲で動く**レンジ（もみ合い）環境**では、誤ったシグナルを発しやすくなります。

* **ADXの役割**: ADXは、価格が上昇しているか下降しているか（方向）ではなく、**トレンドが統計的に存在するかどうか、そしてその強さがどれくらいか**を**0から100の数値**で定量化します。
* **深い推論**: ADXを特徴量として利用することで、モデルは現在の市場環境が「**トレンドフォロー戦略が有効な状態**」なのか、「**平均回帰戦略が有効な状態**」なのかを**自己判断**できるようになり、**戦略のロバスト性**を向上させます。

### 2. ADXの構成要素とその意味

ADXは、以下の3つの主要な要素の組み合わせによって計算されます。

1.  **方向性ムーブメント（DM）の計算**:
    * **+DM**: 現在の高値が前日の高値をどれだけ上回ったか。**買いの圧力**の強さを示す。
    * **-DM**: 前日の安値が現在の安値をどれだけ下回ったか。**売りの圧力**の強さを示す。
    * *注釈*: これらは互いに排他的に計算されます（例えば、+DMが大きい場合、-DMはゼロ）。
2.  **方向性指数（DI）の計算**:
    * **+DI** と **-DI** は、DMを**真の変動幅 (True Range: TR)** で正規化したもので、市場の方向性（買い圧力と売り圧力）を**パーセンテージ**で示します。
3.  **ADXの計算**:
    * **DI** の差（$| +DI - -DI |$）を合計（$+DI + -DI$）で割った **DX** を、さらに平滑化（修正移動平均）したものがADXです。この平滑化により、ADXはトレンドの**持続的な強さ**を示します。

````python
import pandas as pd
import numpy as np
import japanize_matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from arch import arch_model
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# === ヘルパー関数 (GARCH) ===
def calculate_garch_volatility(returns_series, window=100):
    """
    ローリングウィンドウでGARCH(1,1)モデルを適用し、
    1期先の予測ボラティリティ（年率換算ではない）を計算する関数。
    """
    forecasts = {}
    # ループの開始点をウィンドウサイズに設定
    for i in range(window, len(returns_series)):
        # 現在のウィンドウのリターンデータを取得（NaNを削除）
        current_returns = returns_series.iloc[i-window:i].dropna()
        if len(current_returns) < window * 0.8: # データが少なすぎる場合はスキップ
            print("スキップ")
            continue

        try:
            # GARCH(1,1)モデルを定義。rescale=Falseでスケーリングを無効化
            model = arch_model(current_returns * 100, vol='Garch', p=1, q=1, rescale=False)
            # update_freq=0, disp='off' で適合中の出力を抑制
            res = model.fit(update_freq=0, disp='off')
            forecast = res.forecast(horizon=24*1)#1日後のボラティリティを計算
            # 予測された分散から標準偏差（ボラティリティ）を計算
            predicted_vol = np.sqrt(forecast.variance.values[-1,:].sum()) / 100
            forecasts[returns_series.index[i]] = predicted_vol
        except Exception as e:
            # GARCHの収束失敗は頻繁に起こるため、エラーメッセージを抑制
            print(f"GARCH calculation failed at index {i}: {e}")
            pass

    return pd.Series(forecasts)

# === ヘルパー関数 (ADX) ===
def add_adx(df, high_col, low_col, close_col, length=14):#------ADX-------------
    """DataFrameにADXと+DI, -DIを計算して追加する関数（手動計算版）。"""
    high = df[high_col]
    low = df[low_col]
    close = df[close_col]

    prev_high = high.shift(1)
    prev_low = low.shift(1)
    prev_close = close.shift(1)

    # +DM と -DM の計算
    plus_dm = high.diff()
    minus_dm = prev_low - low

    # DMのフィルタリング: +DM > -DM かつ +DM > 0 の場合のみ値を残す
    plus_dm[(plus_dm < 0) | (plus_dm <= minus_dm)] = 0
    # DMのフィルタリング: -DM > +DM かつ -DM > 0 の場合のみ値を残す
    minus_dm[(minus_dm < 0) | (minus_dm <= plus_dm)] = 0

    # TR (True Range) の計算
    tr1 = high - low
    tr2 = abs(high - prev_close)
    tr3 = abs(low - prev_close)
    tr = pd.concat([tr1, tr2, tr3], axis=1).max(axis=1)

    # TR, +DM, -DMの平滑化（Wilder's Smoothing/修正移動平均）
    atr = tr.ewm(alpha=1/length, adjust=False).mean()
    plus_di_smooth = plus_dm.ewm(alpha=1/length, adjust=False).mean()
    minus_di_smooth = minus_dm.ewm(alpha=1/length, adjust=False).mean()

    # +DI, -DIの計算
    plus_di = (plus_di_smooth / (atr + 1e-8)) * 100
    minus_di = (minus_di_smooth / (atr + 1e-8)) * 100

    # DX (Directional Index) の計算
    di_sum = plus_di + minus_di
    di_diff = abs(plus_di - minus_di)
    dx = (di_diff / (di_sum + 1e-8)) * 100

    # ADX (Average Directional Index) の計算 (DXの平滑化)
    adx = dx.ewm(alpha=1/length, adjust=False).mean()

    # DataFrameに列を追加
    # 注: 元のノートブックでは +DI, -DIは最終特徴量リストに含まれていないが、ADX計算のため内部で必要
    df[f'PLUS_DI_{length}'] = plus_di
    df[f'MINUS_DI_{length}'] = minus_di
    df[f'ADX_{length}'] = adx

    return df

# === メインの特徴量作成関数 ===
def create_final_features(df, ma_period=168, garch_window=100):
    """整形済みDataFrameから最終的な特徴量をすべて生成する"""

    df_feat = df.copy()

    eth_close = df_feat["close_eth"]
    btc_close = df_feat["close_btc"]
    eth_ret = eth_close.pct_change()
    btc_ret = btc_close.pct_change()

    # バックテストの add_vol_signals で必要なため ret_1h を作成
    df_feat["ret_1h"] = eth_ret

    # 1. GARCH (vol_eth_garch)
    print("Calculating GARCH... (時間がかかる場合があります)")
    df_feat["vol_eth_garch"] = calculate_garch_volatility(eth_ret, window=garch_window)
    print("GARCH calculation complete.")

    # 2. BTCリターン (ret_btc_1h, ret_btc_24h)
    df_feat["ret_btc_1h"] = btc_ret
    df_feat["ret_btc_24h"] = btc_close.pct_change(24)

    # 3. 価格比率Zスコア (ratio_z_score)
    df_feat["price_ratio_eth_btc"] = eth_close / (btc_close + 1e-8)
    ratio_ma = df_feat["price_ratio_eth_btc"].rolling(ma_period).mean()
    ratio_std = df_feat["price_ratio_eth_btc"].rolling(ma_period).std()
    df_feat["ratio_z_score"] = (df_feat["price_ratio_eth_btc"] - ratio_ma) / (ratio_std + 1e-8)

    # 4. PCAはデータリーケージが発生したため削除2025/12/10

    # RSI
    delta = eth_close.diff()
    gain = np.where(delta > 0, delta, 0)
    loss = np.where(delta < 0, -delta, 0)
    avg_gain = pd.Series(gain, index=df_feat.index).ewm(span=14, adjust=False).mean()
    avg_loss = pd.Series(loss, index=df_feat.index).ewm(span=14, adjust=False).mean()
    rs = avg_gain / (avg_loss + 1e-8)
    df_feat["rsi_eth"] = 100 - (100 / (1 + rs))

    # MACD
    ema12 = eth_close.ewm(span=12).mean()
    ema26 = eth_close.ewm(span=26).mean()
    macd = ema12 - ema26
    signal = macd.ewm(span=9).mean()
    df_feat["macd_hist_eth"] = macd - signal

    # MA Dev & Vol Dist
    df_feat["ma_eth"] = eth_close.rolling(ma_period).mean()
    df_feat["ma_dev_eth"] = (eth_close - df_feat["ma_eth"]) / df_feat["ma_eth"]
    df_feat["vol_dist_eth"] = df_feat["ma_dev_eth"] / df_feat["vol_eth_garch"]


    # 6. 時間特徴量 (hour_sin, hour_cos, day_sin, day_cos)
    df_feat['hour_of_day'] = df_feat.index.hour
    df_feat['day_of_week'] = df_feat.index.dayofweek # 月曜日=0, 日曜日=6
    df_feat['hour_sin'] = np.sin(2 * np.pi * df_feat['hour_of_day'] / 24)
    df_feat['hour_cos'] = np.cos(2 * np.pi * df_feat['hour_of_day'] / 24)
    df_feat['day_sin'] = np.sin(2 * np.pi * df_feat['day_of_week'] / 7)
    df_feat['day_cos'] = np.cos(2 * np.pi * df_feat['day_of_week'] / 7)

    # 7. ADX
    print("Calculating ADX...")
    # NOTE: add_adx は元のDataFrameを直接変更
    df_feat = add_adx(df_feat, high_col='high_eth', low_col='low_eth', close_col='close_eth', length=5)
    df_feat = add_adx(df_feat, high_col='high_eth', low_col='low_eth', close_col='close_eth', length=14)
    df_feat = add_adx(df_feat, high_col='high_eth', low_col='low_eth', close_col='close_eth', length=50)
    df_feat = add_adx(df_feat, high_col='high_eth', low_col='low_eth', close_col='close_eth', length=168)
    print("ADX calculation complete.")

    # 8. 最終クリーンアップ
    # GARCHやローリングのウォームアップ期間で発生したNaNを削除
    df_feat.dropna()

    print("すべての特徴量作成が完了しました。")
    return df_feat


def save_features_to_csv(df_features, output_path="/content/finalfutures.csv"):
    """
    生成された特徴量DataFrameをCSVファイルに出力する。
    """
    if df_features.empty:
        print("DataFrameが空のため、CSV出力はスキップされました。")
        return

    # インデックス（timestamp）を列として含めてCSVに出力
    df_features.to_csv(output_path, index=True)
    print(f"特徴量データが正常に {output_path} に出力されました。")

````

## 📈 ボラティリティ収束則の統計的検証プログラム

本プログラムは、価格が移動平均線から大きく乖離した（高ボラティリティな状態）後に、平均値へ**回帰（Reversion）**する現象、すなわち**ボラティリティ収束則**を統計的に検証することを目的としています。

従来のループ処理を排除し、Pandas/NumPyの**ベクトル化**を利用することで、大量の時系列データに対しても高速かつ効率的に分析を実行できるように最適化されています。

### 1. 動作原理

分析は、以下の2つのシナリオにおける価格の反転（リバーサル）確率を測定します。

#### A. 下落からの反発 (Down-to-Up Reversion)
1.  **条件1（トリガー）**: 価格が**移動平均 (MA)** から、**ボラティリティ (Vol)** を基準とした一定の倍率 $P$ の幅だけ下回った場合。
    $$\text{Close} < \text{MA} - \frac{P}{100} \times \text{Vol} \times \text{MA}$$
2.  **条件2（成功）**: トリガー発生後、$T/X$ 時間以内に価格が $\text{MA}$ から、$\text{Vol}$ を基準とした一定の倍率 $Q$ の幅だけ上回った場合。

#### B. 上昇からの反落 (Up-to-Down Reversion)
1.  **条件1（トリガー）**: 価格が $\text{MA}$ から、$\text{Vol}$ を基準とした一定の倍率 $R$ の幅だけ上回った場合。
    $$\text{Close} > \text{MA} + \frac{R}{100} \times \text{Vol} \times \text{MA}$$
2.  **条件2（成功）**: トリガー発生後、$T/X$ 時間以内に価格が $\text{MA}$ から、$\text{Vol}$ を基準とした一定の倍率 $S$ の幅だけ下回った場合。

### 2. 計算ロジック

* **移動平均 (MA)**: 過去 $\text{MA\_PERIOD}$ 時間の終値の単純移動平均。
* **ボラティリティ (Vol)**: 過去 $\text{VOL\_PERIOD}$ 時間の**対数リターン**の標準偏差。これは価格スケールに変換するため、MAを乗算して使用されます（例：$\text{Vol} \times \text{MA}$）。

#### 💡 ベクトル化による効率的な成功判定
最も重要な最適化ポイントは、トリガー後の成功判定です。元のプログラムで使われていた非効率な `for` ループとスライシングの代わりに、以下のベクトル化テクニックを使用しています。

1.  データフレームを**逆順**にし、$\text{T/X}$ 期間の**ローリング最大値/最小値**を計算します。
2.  計算結果を再度**逆順**に戻し、`shift(-1)`で1期間シフトすることで、各時点 $t$ における**将来 $T/X$ 期間内の最大値/最小値**（$\text{future\_max}$ / $\text{future\_min}$）を効率的に算出しています。
3.  トリガー条件と、$\text{future\_max}$ / $\text{future\_min}$ が成功条件を満たすかをブール演算（AND）で一括判定し、成功回数を集計します。

### 3. パラメータ設定

以下のパラメータを変更することで、検証条件を調整できます。

| 変数名 | 定義 | 説明 |
| :--- | :--- | :--- |
| `MA_PERIOD` | 168h (7日) | 移動平均線の計算期間（時間）。 |
| `VOL_PERIOD` | 168h (7日) | ボラティリティ（対数リターン標準偏差）の計算期間（時間）。 |
| **$P$** | 80% | 下落トリガーの閾値（$\text{MA} - P\% \times \text{Vol} \times \text{MA}$）。 |
| **$Q$** | 80% | 反発成功の閾値（$\text{MA} + Q\% \times \text{Vol} \times \text{MA}$）。 |
| **$R$** | 200% | 上昇トリガーの閾値（$\text{MA} + R\% \times \text{Vol} \times \text{MA}$）。 |
| **$S$** | 100% | 反落成功の閾値（$\text{MA} - S\% \times \text{Vol} \times \text{MA}$）。 |
| **$X$** | 1 | 検証期間 $T$ を $T/X$ に分割する係数。$\text{T\_periods}$ で指定された期間 $T$ 全体で成功判定を行う場合は $X=1$ とします。 |

````python
import pandas as pd
import numpy as np

# --- パラメータ設定 ---
MA_PERIOD = 24 * 7  # 例: 7日間 (168時間)
VOL_PERIOD = 24 * 7 # 例: 7日間 (168時間)

# 下落からの反発確率の検証パラメータ
P = 80  # P値 (%)
Q = 80  # Q値 (%)

# 上昇からの反落確率の検証パラメータ
R = 200 # R値 (%)
S = 100 # S値 (%)

# 検証期間Tを何分割して条件2の発生を待つか
X = 1


def calculate_indicators(df_close, ma_period, vol_period):
    """
    移動平均とボラティリティを計算し、必要なデータフレームを準備する。
    """
    df = pd.DataFrame({'close': df_close.copy()})

    # 移動平均 (t-1までのデータ)
    df['ma'] = df['close'].shift(1).rolling(window=ma_period).mean()

    # 1時間ごとの対数リターンを計算
    df['log_return'] = np.log(df['close'] / df['close'].shift(1))

    # ボラティリティ（対数リターンの標準偏差） (t-1までのデータ)
    # Volはリターンの標準偏差なので、価格スケールに合わせる係数 (ma) を事前に計算
    df['volatility'] = df['log_return'].shift(1).rolling(window=vol_period).std()

    # 計算初期のNaNを削除
    df.dropna()

    return df


def analyze_reversion_probability_optimized(df_raw, periods, ma_period, vol_period, p, q, r, s, x):
    """
    指定された条件に基づき、価格の反転確率をベクトル化で計算する関数（最適化版）
    """
    # データの準備（df_raw["close_eth"]を想定）
    if "close_eth" not in df_raw:
        print("エラー: df_rawに 'close_eth' カラムが見つかりません。")
        return

    df = calculate_indicators(df_raw["close_eth"], ma_period, vol_period)

    print("\n--- 分析開始 (ベクトル化最適化版) ---")

    # 条件1/条件2の基準値を事前に計算
    df['down_condition_1_limit'] = df['ma'] - (p / 100) * df['volatility'] * df['ma']
    df['up_condition_2_limit'] = df['ma'] + (q / 100) * df['volatility'] * df['ma']
    df['up_condition_1_limit'] = df['ma'] + (r / 100) * df['volatility'] * df['ma']
    df['down_condition_2_limit'] = df['ma'] - (s / 100) * df['volatility'] * df['ma']

    for name, t_hours in periods.items():
        lookahead_window = int(t_hours / x)
        if lookahead_window < 1:
            print(f"\n期間 T={name}({t_hours}h), X={x}: lookahead_windowが1未満のためスキップします。")
            continue

        print(f"\n--- 検証期間 T = {name} ({t_hours}時間), チェック区間 T/X = {lookahead_window}時間 ---")

        # ----------------------------------------------------
        # 💡 ベクトル化のコア技術: 将来の期間 T/X 内の最大値/最小値を計算
        # ----------------------------------------------------
        # 逆順にしてrolling().max()/min()を適用し、再度逆順に戻すことで「将来のローリング最大値/最小値」を計算

        # 将来の期間T/Xにおけるcloseの最大値
        future_max = df['close'][::-1].rolling(window=lookahead_window, min_periods=1).max()[::-1].shift(-1)
        # 将来の期間T/Xにおけるcloseの最小値
        future_min = df['close'][::-1].rolling(window=lookahead_window, min_periods=1).min()[::-1].shift(-1)

        # NaNを考慮してインデックスを調整
        df['future_max'] = future_max.reindex(df.index)
        df['future_min'] = future_min.reindex(df.index)

        # --- 下落からの反発確率 (Down -> Up) ---

        # 条件1のトリガーを判定
        down_triggers = df['close'] < df['down_condition_1_limit']
        down_trigger_count = down_triggers.sum()

        if down_trigger_count > 0:
            # 条件2の成功をベクトル化で判定
            # トリガーが発生した行に絞り込み、将来の最大値が条件2の基準を超えたかチェック
            down_to_up_success = (
                (down_triggers) &
                (df['future_max'] > df['up_condition_2_limit'])
            )
            down_to_up_success_count = down_to_up_success.sum()

            down_to_up_prob = (down_to_up_success_count / down_trigger_count) * 100
        else:
            down_to_up_prob = 0
            down_to_up_success_count = 0


        print(f"▼ 下落からの反発確率:")
        print(f"    条件1発生回数 (価格 < MA - {p}%*Vol): {down_trigger_count} 回")
        print(f"    条件2成功回数 (その後価格 > MA + {q}%*Vol): {down_to_up_success_count} 回")
        print(f"    => 確率: {down_to_up_prob:.2f} %")


        # --- 上昇からの反落確率 (Up -> Down) ---

        # 条件1のトリガーを判定
        up_triggers = df['close'] > df['up_condition_1_limit']
        up_trigger_count = up_triggers.sum()

        if up_trigger_count > 0:
            # 条件2の成功をベクトル化で判定
            # トリガーが発生した行に絞り込み、将来の最小値が条件2の基準を下回ったかチェック
            up_to_down_success = (
                (up_triggers) &
                (df['future_min'] < df['down_condition_2_limit'])
            )
            up_to_down_success_count = up_to_down_success.sum()

            up_to_down_prob = (up_to_down_success_count / up_trigger_count) * 100
        else:
            up_to_down_prob = 0
            up_to_down_success_count = 0

        print(f"▲ 上昇からの反落確率:")
        print(f"    条件1発生回数 (価格 > MA + {r}%*Vol): {up_trigger_count} 回")
        print(f"    条件2成功回数 (その後価格 < MA - {s}%*Vol): {up_to_down_success_count} 回")
        print(f"    => 確率: {up_to_down_prob:.2f} %")


if __name__ == '__main__':


    if df_raw.get("close_eth") is not None:
        # 検証する期間Tを定義（単位：時間）
        T_periods = {
            '1日': 24,
            '1週間': 24 * 7,
            '1ヶ月': 24 * 30, # 元の14日から一般的な30日に変更
            '1年間': 24 * 365 # 元の30日から一般的な365日に変更
        }

        # 分析を実行
        analyze_reversion_probability_optimized(
            df_raw=df_raw,
            periods=T_periods,
            ma_period=MA_PERIOD,
            vol_period=VOL_PERIOD,
            p=P, q=Q, r=R, s=S, x=X
        )
    else:
        print("エラー: 分析に必要なデータ (df_raw['close_eth']) が見つからないか、不適切な形式です。")
````

## 📖 プロジェクト解説：AIと市場心理を融合させたETHトレーディングモデル


これは、仮想通貨イーサリアム（ETH）の価格変動を予測し、自動で売買するための**アルゴリズムトレード・モデル**です。

このモデルの最大の特徴は、**2つの異なる戦略**を賢く組み合わせている点にあります。

1.  🧠 **AI（XGBoost）による精密予測**：
    * AIが過去のデータから「将来（11時間後）価格が上がるか、上がらないか」を学習し、予測します。
    * 特に、価格が**上がらない（＝横ばい、または下落する）局面**を高い精度で予測することに重点を置いています。これにより、不要な損失を避けることを目指します。
2.  ⚖️ **ボラティリティ収束則（平均回帰）の利用**：
    * 市場には「価格が行き過ぎたら、いずれ平均に戻ってくる」という性質があります。
    * 価格が統計的に見て「売られすぎ」なレベルまで下落したらAIの予測に加えて「買い」の判断を強め、「買われすぎ」なレベルまで上昇したら「売り」の判断を強めます。

この2つを組み合わせることで、**「AIによる冷静な予測」**と**「市場の行き過ぎた心理（ボラティリティ）を捉える逆張り」**の両方の長所を活かした、堅牢な戦略を構築しています。

### ⚙️ プログラムの仕組み（セクション別）

このプログラムがどのような流れで動いているのか、上から順に見ていきましょう。

#### 1. `import` (ライブラリの準備)
まず、分析に必要な「道具」を揃えます。
* `pandas`や`numpy`：データを表計算ソフトのように扱ったり、数値計算をしたりするための基本ツールです。
* `xgboost`：今回の主役である**XGBClassifier**という強力なAI（機械学習モデル）です。
* `matplotlib`：結果をグラフで表示するために使います。

#### 2. `=== 設定 ===` (戦略の設計図)
ここでは、モデルの「性格」や「ルール」を細かく決めています。

* `final_feature_columns`:
    * AIが未来を予測するために使う「ヒント（**特徴量**）」のリストです。
    * `volume_eth`（取引量）や`ret_btc_1h`（ビットコインの収益率）などに加え、`ADX_5`, `ADX_50`, `ADX_168`といった**トレンドの強さ**を示すテクニカル指標を追加し、予測精度向上を図っています。
* `HORIZON = 11`:
    * 「**11時間後**」の価格を予測対象にします。
* `CLASSIFICATION_THRESHOLD = 0.003`:
    * 11時間後に価格が**0.3%**より大きく上昇した場合のみを「上昇」と定義します。これ未満の小さな動きは「非上昇（横ばい・下落）」とみなします。
* `XGB_PARAMS`:
    * AI (XGBoost) の学習方法を細かくチューニングする設定です。これにより、モデルが賢く学習できるように調整しています。
* `VOL_WEIGHT = 0.3`:
    * 最終的な売買判断において、AIの予測と「ボラティリティ収束シグナル」の**重要度（重み）**を調整します。ここでは、ボラティリティ収束シグナルの影響度を30%に設定しています。
* `UP_HITS_REQUIRED` / `DOWN_STREAK_REQUIRED`:
    * 「ダマシ（ノイズ）」を避けるためのルールです。
    * 「買い」は、シグナルが直近12時間で**9回**発生したら実行します（確実性を重視）。
    * 「売り」は、シグナルが**3回連続**で発生したら実行します（危険を素早く察知）。

#### 3. `=== データ読み込み ===` (過去データの整理)
`finalfutures.csv`というファイルから過去の価格データを読み込み、時系列（時間の流れ）に沿ってデータを並べ替え、分析しやすい形に整えます。

#### 4. `=== ターゲット作成 ===` (AIへの「正解」の教育)
AIに何を予測させるかを定義する、非常に重要な部分です。

* `future_ret`で11時間後の価格変動率を計算します。
* 変動率が`0.3%`より大きければ「**1 (上昇)**」
* `0.3%`以下（横ばいや下落）なら「**0 (非上昇)**」
* という「正解ラベル（`y`）」を作成します。
* **このモデルの核心**は、AIがこの「**0 (非上昇)**」を正確に予測することです。これにより、「今は買うべきではない（＝下落リスクがある）」という判断が可能になります。

#### 5. `=== Walk-forward 学習 ===` (AIの賢い学習法)
金融データでAIを学習させるための特別な方法、「**ウォークフォワード学習**」を行います。

* 通常のAI学習では全データを一度に学習させますが、金融市場は常に変化しています。
* そこで、「過去のデータだけ」を使って学習し、直近の未来を予測します。
* 予測が終わったら、その「直近の未来」のデータも学習データに加え、再度学習し直します。
* これを繰り返すことで、**市場の変化にAIが適応し続ける**ことができます。
* ここでは、`RETRAIN_EVERY_N = 1000`（約1000時間ごと）にAIを再学習させています。

#### 6. `=== ボラティリティ収束シグナル ===` (戦略の柱①：逆張り)
このモデルの戦略の1つ目の柱、「ボラティリティ収束（平均回帰）」を計算します。

* `ma_vol`（移動平均）：市場の長期的な「平均価格」を計算します。
* `vol_sigma`（標準偏差）：価格の「荒れ具合（ボラティリティ）」を計算します。
* `reb_signal`（買いシグナル）：現在の価格が「平均価格」から「荒れ具合」を考慮しても**異常に安く（売られすぎ）**なった場合に「1」が立ちます。
* `rev_signal`（売りシグナル）：現在の価格が**異常に高く（買われすぎ）**なった場合に「1」が立ちます。

#### 7. `=== バックテスト ===` (戦略のシミュレーション)
いよいよ、2つの戦略を組み合わせて、過去のデータでシミュレーション（バックテスト）を行います。

* `comp_up`（総合的な「買い」判断）:
    * `AIの上昇予測` ＋ (`ボラティリティの買いシグナル` × `VOL_WEIGHT`)
    * AIが「上がる」と予測し、かつ市場が「売られすぎ」の場合に、より強く「買い」と判断します。
* `comp_down`（総合的な「売り」判断）:
    * `AIの非上昇予測` ＋ (`ボラティリティの売りシグナル` × `VOL_WEIGHT`)
    * **ここがポイントです。** AIが「上がらない（下落/横ばい）」と予測し、かつ市場が「買われすぎ」の場合に、より強く「売り（ポジション解消）」と判断します。
* このロジックに従い、初期資金`INITIAL_EQUITY`（100万円）が時間とともにどう増減するかをシミュレーションします。

#### 8. `=== 指標 ===` (成績の評価)
シミュレーション結果を客観的に評価するための指標を計算します。

* `CAGR` (年率リターン)：1年あたり平均何％の利益が出たか。
* `MaxDD` (最大ドローダウン)：最も運が悪かった時期に、資産が最大何％減少したか（リスクの大きさ）。
* `Sharpe` (シャープレシオ)：リスクに対してどれだけ効率よくリターンを上げたか（数値が高いほど優秀）。

#### 9. `=== メイン ===` (全体の実行と結果表示)
これまでの全ての処理を順番に実行し、結果を表示します。

* `Classification Report` (分類レポート):
    * AIが「1 (上昇)」と「**0 (非上昇)**」をどれだけ正確に予測できたかを示す成績表です。
    * 特に「0」に対する **`precision` (適合率)** や **`recall` (再現率)** を見ることで、**このモデルが重視する「下落局面の予測精度」**を具体的に評価できます。
* 最後に、この戦略の資産推移グラフと、単純に買って持ち続けた場合（Buy&Hold）のグラフを比較表示します。

---

### 🚀 この戦略の強みと未来

このモデルは、単に「上がる」と予測するだけでなく、「**上がらない（下落する）**」局面をAIで正確に察知すること、そして市場の**行き過ぎた価格変動（ボラティリティ）**を利用することを組み合わせた、現実的で堅牢な戦略を目指しています。

AIの予測と古典的な市場の性質を融合させることで、より安定したパフォーマンスを追求しています。


#分類モデル×ボラティリティシグナルモデル(ボラティリティシグナル:=0.4)

````python
#XGboost×ボラティリティシグナ ADX5,50,168追加

import os
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import deque

from sklearn.metrics import classification_report, confusion_matrix
from xgboost import XGBClassifier

# === 設定 ===
CSV_PATH = "/content/finalfutures.csv"
DATETIME_COLUMN = "timestamp"
PRICE_COLUMN = "close_eth"
TIMEZONE = None

# === 修正後の特徴量リスト（これをXGBoostの学習に使用します） ===
final_feature_columns = [
    # 1. 基本データ
    'volume_eth', 'volume_btc', 'vol_eth_garch', 'ret_btc_1h', 'ret_btc_24h',
    'ratio_z_score',

    # 2. 時間的特徴量
    'hour_sin', 'hour_cos', 'day_sin', 'day_cos',

    # 3. トレンド/環境認識 (ADX)
    'ADX_14', 'ADX_5', "ADX_50", "ADX_168",

    # 4. 【NEW】PCAの代わりに追加したテクニカル指標
    'ma_dev_eth',      # 移動平均乖離率 (買われすぎ判定)
    'vol_dist_eth',    # ボラティリティ距離 (異常値判定)
    'rsi_eth',         # RSI (逆張り判定)
    'macd_hist_eth'    # MACDヒストグラム (トレンド転換判定)
]

HORIZON = 11
CLASSIFICATION_THRESHOLD = 0.003

#XGB_PARAMS = dict(
    #learning_rate = 0.05,
    #n_estimators = 400,
    #max_depth = 4,
    #subsample = 0.7,
    #colsample_bytree = 0.6,
    #min_child_weight = 3,
    #gamma = 0.1,
    #reg_alpha = 0.1,
    #reg_lambda = 1.0,
#)

XGB_PARAMS = dict(
    learning_rate = 0.02,
    n_estimators = 800,
    max_depth = 6,
    subsample = 0.8,
    colsample_bytree = 0.8,
    min_child_weight = 1,
    gamma = 0.0,
    reg_alpha = 0.0,
    reg_lambda = 1.0

)


COMMISSION = 0.0005
SLIPPAGE = 0.0
USE_PORTFOLIO_95 = True

SIGNAL_LEN = 12
UP_PROB_THRESH = 0.7
UP_HITS_REQUIRED = 9
DOWN_PROB_THRESH = 0.65
DOWN_STREAK_REQUIRED = 3

FEATURE_BUFFER_HOURS = 600
TEST_SIZE_FRACTION = 0.25
RETRAIN_EVERY_N = 1000
MIN_TRAIN_SIZE = 8000
INITIAL_EQUITY = 1_000_000.0

VOL_WEIGHT = 0.4  # ボラティリティ収束シグナル重み

# === データ読み込み ===
def load_data(path):
    df = pd.read_csv(path)
    df[DATETIME_COLUMN] = pd.to_datetime(df[DATETIME_COLUMN])
    if TIMEZONE:
        if df[DATETIME_COLUMN].dt.tz is None:
            df[DATETIME_COLUMN] = df[DATETIME_COLUMN].dt.tz_localize(TIMEZONE)
        else:
            df[DATETIME_COLUMN] = df[DATETIME_COLUMN].dt.tz_convert(TIMEZONE)
    df = df.sort_values(DATETIME_COLUMN).set_index(DATETIME_COLUMN)
    if "ret_1h" not in df.columns:
        df["ret_1h"] = df[PRICE_COLUMN].pct_change()
    return df

# === ターゲット作成 (二値分類に変更) ===
def make_target(df):
    future_ret = df[PRICE_COLUMN].shift(-HORIZON) / df[PRICE_COLUMN] - 1.0
    y = pd.Series(index=df.index, dtype='float64')
    y[future_ret > CLASSIFICATION_THRESHOLD] = 1.0  # 上昇
    y[future_ret <= CLASSIFICATION_THRESHOLD] = 0.0 # 非上昇 (横ばい or 下落)
    return y

# === Walk-forward 学習 (二値分類に変更) ===
def walk_forward_proba(df, X_cols, y):
    X_cols = [c for c in X_cols if c in df.columns]
    df2 = df.iloc[FEATURE_BUFFER_HOURS:].copy()
    y2 = y.loc[df2.index]
    X_all = df2[X_cols]
    n = len(df2)
    test_start = max(int(n*(1-TEST_SIZE_FRACTION)), MIN_TRAIN_SIZE)
    test_idx = X_all.index[test_start:]
    proba_records = []
    model = None
    last_fit_i = None
    for i, ts in enumerate(test_idx):
        if (model is None) or (last_fit_i is None) or (i - last_fit_i >= RETRAIN_EVERY_N):
            end_loc = test_start + i
            X_train = X_all.iloc[:end_loc]
            y_train = y2.iloc[:end_loc]
            valid_idx = X_train.dropna().index.intersection(y_train.dropna().index)
            if len(valid_idx) >= 20 and len(y_train.loc[valid_idx].unique()) > 1:
                model = XGBClassifier(**XGB_PARAMS)
                model.fit(X_train.loc[valid_idx].values, y_train.loc[valid_idx].values)
            else:
                model = None
            last_fit_i = i
        x_row = X_all.loc[[ts]].values
        if model is None:
            p = np.array([[0.5, 0.5]]) # 2クラス用のデフォルト確率
        else:
            p = model.predict_proba(x_row)
        proba_records.append((ts,float(p[0,0]),float(p[0,1])))
    # 2クラス用のDataFrameを作成
    proba_df = pd.DataFrame(proba_records, columns=["timestamp","p_not_up","p_up"]).set_index("timestamp")
    y_true = y2.loc[proba_df.index]
    return proba_df, y_true

# === ボラティリティ収束シグナル (変更なし) ===
def add_vol_signals(df, ma_period=24*7, vol_period=24*7, P=0.8, R=2.0):
    d = df.copy()
    d["ma_vol"] = d[PRICE_COLUMN].rolling(ma_period).mean()
    d["vol_sigma"] = d["ret_1h"].rolling(vol_period).std()
    d["vol_sigma"].replace(0,np.nan)
    d["reb_signal"] = ((d[PRICE_COLUMN] < d["ma_vol"] - P * d["vol_sigma"])).astype(int).fillna(0)
    d["rev_signal"] = ((d[PRICE_COLUMN] > d["ma_vol"] + R * d["vol_sigma"])).astype(int).fillna(0)
    return d[["ma_vol","vol_sigma","reb_signal","rev_signal"]]

# === バックテスト (二値分類に変更) ===
def backtest_composite(df, proba_df):
    df_sub = df.loc[proba_df.index].copy()
    vol_df = add_vol_signals(df_sub)
    df_sub = df_sub.join(vol_df)
    prices = df_sub[PRICE_COLUMN]
    fill_prices = prices.shift(-1)
    cash, pos = INITIAL_EQUITY, 0.0
    equity = pd.Series(index=proba_df.index,dtype=float)
    logs = []
    up_hits = deque(maxlen=SIGNAL_LEN)
    down_streak = 0

    comp_up = np.clip(proba_df["p_up"] + VOL_WEIGHT * df_sub["reb_signal"],0,1)
    # p_down の代わりに p_not_up を使用
    comp_down = np.clip(proba_df["p_not_up"] + VOL_WEIGHT * df_sub["rev_signal"],0,1)

    for i, t in enumerate(proba_df.index):
        p_up_c = float(comp_up.iloc[i])
        p_down_c = float(comp_down.iloc[i])
        up_hits.append(p_up_c >= UP_PROB_THRESH)
        down_streak = down_streak + 1 if p_down_c >= DOWN_PROB_THRESH else 0

        price = prices.loc[t]
        next_price = fill_prices.loc[t] if t in fill_prices.index else np.nan
        if np.isnan(next_price):
            equity.loc[t] = cash + pos*price
            continue
        if (sum(up_hits)>=UP_HITS_REQUIRED) and pos==0:
            use_ratio = 0.95 if USE_PORTFOLIO_95 else 1.0
            notional = cash*use_ratio
            fill = next_price*(1+SLIPPAGE)
            qty = (notional*(1-COMMISSION))/fill
            cash -= qty*fill; pos += qty
            logs.append({"timestamp":t,"side":"BUY","price":float(fill),"qty":float(qty),"cash_after":float(cash)})
        elif (down_streak>=DOWN_STREAK_REQUIRED) and pos>0:
            fill = next_price*(1-SLIPPAGE)
            cash += pos*fill*(1-COMMISSION)
            logs.append({"timestamp":t,"side":"SELL","price":float(fill),"qty":float(pos),"cash_after":float(cash)})
            pos=0
        equity.loc[t] = cash + pos*price

    # Buy&Hold
    bh_entry = prices.iloc[0]*(1+SLIPPAGE)
    bh_qty = (INITIAL_EQUITY*(1-COMMISSION))/bh_entry
    bh_equity = bh_qty*prices

    trades_df = pd.DataFrame(logs).set_index("timestamp") if logs else pd.DataFrame(columns=["side","price","qty","cash_after"])
    return equity, bh_equity, trades_df

# === 指標 (変更なし) ===
def metrics(equity):
    eq = equity.dropna()
    rets = eq.pct_change().dropna()
    periods_per_year = 24*365
    cagr = (eq.iloc[-1]/eq.iloc[0])**(periods_per_year/len(eq))-1
    maxdd = (1 - eq/eq.cummax()).max()
    sharpe = (rets.mean()/(rets.std()+1e-12))*math.sqrt(periods_per_year)
    return {"CAGR": cagr, "MaxDD": maxdd, "Sharpe": sharpe}

# === メイン (二値分類に変更) ===
def main():
    df = load_data(CSV_PATH)
    y = make_target(df)
    valid_idx = df.index.intersection(df[final_feature_columns].dropna().index).intersection(y.dropna().index)
    df_clean = df.loc[valid_idx]
    X_clean = df_clean[final_feature_columns]
    y_clean = y.loc[valid_idx]

    proba_df, y_true = walk_forward_proba(df_clean, final_feature_columns, y_clean)

    # Classification report (二値分類用に変更)
    # 確率0.5を閾値として予測ラベルを生成
    y_hat = (proba_df['p_up'] > 0.5).astype(int)
    print("\n=== Classification Report (0: Not Up, 1: Up) ===")
    print(classification_report(y_true.values, y_hat, digits=4))
    print("Confusion Matrix:\n", confusion_matrix(y_true.values, y_hat))

    equity, bh_equity, trades_df = backtest_composite(df_clean, proba_df)

    print("\nStrategy:", {k:round(v,4) for k,v in metrics(equity).items()})
    print("Buy&Hold:", {k:round(v,4) for k,v in metrics(bh_equity).items()})

    # plot
    plt.figure(figsize=(12,5))
    plt.plot(equity,label="Strategy")
    plt.plot(bh_equity,label="Buy&Hold")
    plt.legend(); plt.grid(True,alpha=0.3); plt.show()

    # save trades
    trades_df.to_csv(os.path.splitext(CSV_PATH)[0]+"_trades.csv")

if __name__=="__main__":
    main()
````
